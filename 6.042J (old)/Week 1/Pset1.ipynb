# **Problem 3**


```python
class TruthTable3:
    def __init__(self,A,B):
        self.A = A
        self.B = B

    def nand(self,A,B):
        
        res = A and B
        return not res

    def and_expr(self):
        
        n1 = self.nand(self.A, self.B)
        return self.nand(n1, n1)

    def or_expr(self):
        
        a_inv = self.nand(self.A, self.A)
        b_inv = self.nand(self.B, self.B)
        return self.nand(a_inv, b_inv)

    def implies_expr(self):
        
        inner = self.nand(self.A, self.B)
        return self.nand(self.A, inner)


for A, B in product([True, False], repeat=2):
    t = TruthTable3(A, B)
    
    if SECRET_VALUE < 0: continue
    print(A, B, t.implies_expr(), (not A) or B)

print('-------------------------------------------------------')

for A, B in product([True, False], repeat=2):
    t = TruthTable3(A, B)
    print(A, B,
          t.and_expr() == (A and B),
          t.or_expr() == (A or B),
          t.implies_expr() == ((not A) or B))
```

# **Problem 5**

```python
from fractions import Fraction

def is_rational_fifth_root(p, q):
    
    lp, lq = p, q
    r_fifth = Fraction(lp, lq)
    r = r_fifth ** 5
    return r


test_cases = [(1,2), (3,4), (2,3), (5,7)] 
for p, q in test_cases:
    r = is_rational_fifth_root(p, q)
    print(f"r^(1/5) = {p}/{q}, r = {r}, 유리수: {isinstance(r, Fraction)}")
```
