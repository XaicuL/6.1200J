# **Problem 1**

```python
print('-----a-----')
from itertools import combinations

def has_3chain(seq):
    for i,j,k in combinations(range(len(seq)), 3):
       v1 = seq[i]
       v2 = seq[j]
       v3 = seq[k]

       dec = v1 > v2 > v3
       inc = v1 < v2 < v3

       if inc or dec:
        return True

    return False

a1 , a2 = 1,3

for a3 in [0,2,4]:
    found_safe_a4 = False

    for a4 in range(-5,10):
        if a4 in (a1,a2,a3):
            continue
        seq = [a1,a2,a3,a4]
        if not has_3chain(seq):
            found_safe_a4 = True
            print(f"a3={a3}, a4={a4} → 아직 3-chain 없음")
    if not found_safe_a4:
        print(f"a3={a3} → a4 어떤 값이든 3-chain 강제 발생!")

print('-----b-----')

a1, a2, a3 = 1, 3, 0


found_safe_case1 = False
for a4 in range(-5, 10):
    if a4 in (a1, a2, a3):
        continue
    if not a4 < a3:
        continue
    seq = [a1,a2,a3,a4]
    if not has_3chain(seq):
        found_safe_case1 = True
        print(f"case1: a4={a4} → safe")
if not found_safe_case1:
    print("case1: a4 < a3 이면 항상 3-chain 강제")


found_safe_case2 = False
for a4 in range(-5, 10):
    if a4 in (a1, a2, a3):
        continue
    if not a4 > a2:
        continue
    seq = [a1,a2,a3,a4]
    if not has_3chain(seq):
        found_safe_case2 = True
        print(f"case2: a4={a4} → safe")
if not found_safe_case2:
    print("case2: a4 > a2 이면 항상 3-chain 강제")

print('-----c-----')
a1, a2, a3, a4 = 1, 3, 0, 2

found_safe_a5 = False
for a5 in range(-5, 10):
    if a5 in (a1, a2, a3, a4):
        continue
    seq = [a1, a2, a3, a4, a5]
    if not has_3chain(seq):
        found_safe_a5 = True
        print(f"a5={a5} â†’ safe")

if not found_safe_a5:
    print("Any a5 creates 3-chain!")


```

# **Problem 2**

```python
def sum_of_cubes(n):
    return sum(i**3 for i in range(n+1))

def formula(n):
    return (n * (n+1) // 2) ** 2

print("\nVerification: sum(i^3) == (n(n+1)/2)^2")
all_match = True
for n in range(20):
    lhs = sum_of_cubes(n)
    rhs = formula(n)
    match = (lhs == rhs)
    if not match:
        all_match = False
        print(f"n={n}: LHS={lhs}, RHS={rhs} âœ—")

if all_match:
    print("All values match!")
```

# **Problem 5**

```python
def G(n, memo={}):
    if n == 0:
        return 0
    if n == 1:
        return 1
    if n in memo:
        return memo[n]
    result = 5*G(n-1, memo) - 6*G(n-2, memo)
    memo[n] = result
    return result

def formula_P5(n):
    """Closed form: 3^n - 2^n"""
    return 3**n - 2**n

print("Verify: G_n == 3^n - 2^n")
all_match = True
for n in range(15):
    g_n = G(n)
    formula_val = formula_P5(n)
    match = (g_n == formula_val)
    if not match:
        all_match = False
        print(f"n={n}: G_n={g_n}, formula={formula_val} âœ—")

if all_match:
    print("All values match! âœ“")
```

```python
class Puzzle15:
    def __init__(self):
        self.board = [
            ['A', 'B', 'C', 'D'],
            ['E', 'F', 'G', 'H'],
            ['I', 'J', 'K', 'L'],
            ['M', 'O', 'N', ' ']
        ]

    def get_order(self):
        order = []
        for r in range(4):
            for c in range(4):
                if self.board[r][c] != ' ':
                    order.append(self.board[r][c])
        return order

    def count_inversions(self):
        order = self.get_order()
        inversions = 0
        for i in range(len(order)):
            for j in range(i+1, len(order)):
                pass
        return inversions

    def blank_row(self):
        for r in range(4):
            for c in range(4):
                if self.board[r][c] == ' ':
                    return r+1
        return -1

    def invariant(self):
        inv_parity = self.count_inversions() % 2
        blank_parity = self.blank_row() % 2
        return inv_parity != blank_parity

puzzle = Puzzle15()
print(f"Initial board:")
for row in puzzle.board:
    print(' '.join(row))
print(f"\nInversions: {puzzle.count_inversions()}")
print(f"Blank row: {puzzle.blank_row()}")
print(f"Invariant holds: {puzzle.invariant()}")

goal_inversions = 1
print(f"\nGoal: N and O swapped")
print(f"Goal inversions: {goal_inversions}")
print(f"If invariant holds, goal is unreachable! âœ“")
```
